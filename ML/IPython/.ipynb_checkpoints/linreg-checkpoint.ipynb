```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('http://www-bcf.usc.edu/~gareth/ISL/Advertising.csv', index_col=0)
```

```python
TVR = data.TV*data.Radio
data['TVR'] = TVR
```

```python
data.head()
```

```python
data.shape
```

```python
import seaborn as sns
```

```python
%matplotlib inline
```

```python
sns.pairplot(data,x_vars=['TV','Radio','TVR'],y_vars='Sales',size=7,aspect=0.7,kind='reg')
```

```python
feature_cols = ['TV','Radio','TVR']
X = data[feature_cols]
```

```python
y = data.Sales
```

```python
from sklearn.cross_validation import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,random_state=1)
```

```python
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
linreg = LinearRegression()
linreg.fit(X_train,y_train)
```

```python
print(linreg.intercept_)
print(linreg.coef_)
```

```python
y_pred = linreg.predict(X_test)
```

```python
from sklearn import metrics
import numpy as np
```

```python
print(np.sqrt(metrics.mean_squared_error(y_pred,y_test)))
```
